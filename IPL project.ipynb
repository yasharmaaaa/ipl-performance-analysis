# Statistical Analysis of IPL Performance: Kohli vc Sharma

## Introduction
This project present a statistical analysis of the IPL performanceof two of the most prominent Indian cricketers,Virat Kohli and Rohit Sharma.The objective is to campare their performance based on various metrics such as total runs,consistency, and seasonal performance.

Using data visualization techniques in python, this analysis highlights trends and key differences in their batting performance over the years in the Indian Premier League (IPL).

## Dataset Description

The dataset used in this project contains season-wise performance data of two prominent IPL players, Virat Kohli and Rohit Sharma. It includes information such as the number of runs scored by each player across different IPL seasons.

Additionally, other datasets are used to analyze overall batting statistics, including metrics like average, strike rate, and boundary counts. These datasets help in performing a comprehensive comparison between the two players.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sharma-kohli.csv')
df.head()
```

```python
df = pd.read_csv('sharma-kohli.csv')
plt.figure(figsize=(12,4))
plt.plot(df['index'],df['RG Sharma'],label='R Sharma',marker='o')
plt.plot(df['index'],df['V Kohli'],label='V Kohli',color='r',marker='s')
plt.title('IPL Performance Comparison:Rohit Sharma vc Virat kohli')
plt.xlabel('season')
plt.ylabel('runs')
#legend
plt.legend()
plt.grid(True)
plt.show()
```

```python
total_rohit = df['RG Sharma'].sum()
total_kohli = df['V Kohli'].sum()

print('Total Runs of Rohit Sharma:',total_rohit)
print('Total Runs of Virat Kohli:',total_kohli)
```

```python
players = ['Rohit Sharma','Virat Kohli']
runs = [total_rohit,total_kohli]

plt.figure(figsize=(6,5))
plt.bar(players,runs,color=['#42a4f5','red'])

plt.title('Total Runs Comparison')
plt.ylabel('Runs')

plt.show()
```

```python
print('Rohit Best Season Runs:',df['RG Sharma'].max())
print('Kohli Best Season Runs:',df['V Kohli'].max())
```

## Conclusion

From the above analysis, it can be observed that both players have shown remarkable performance in the IPL. Virat Kohli has demonstrated higher consistency and peak performance in certain seasons, while Rohit Sharma has maintained steady contributions across multiple years.

Overall, both players have significantly contributed to their teams, making them two of the most successful batsmen in IPL history.
